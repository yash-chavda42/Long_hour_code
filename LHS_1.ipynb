```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_long=pd.read_csv("Arjun_Assignment_data/dataset1/posts_long.csv")
dataset_long
```

```python
#task1 find no of tag per question
tag_long=dataset_long.values[:,9]
taglong2=list()
tagsl=list()
mergelist=list()
for i in range(len(tag_long)):
  x = tag_long[i].split("<")    
  s = "".join(x)
  x = s.split(">")
  x.pop()
  taglong2.append(x)
  tagsl.append(len(x)) 
  for j in x:
    mergelist.append(j)
```

```python
print(len(mergelist))
```

```python
print(tagsl)
```

```python
frequency = {}

# iterating over the list
for item in tagsl:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
print(frequency)
```

```python
courses = list(frequency.keys())
values = list(frequency.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("No Of Tags")
plt.ylabel("Frequncy")
plt.show()
```

```python
x=np.array(mergelist)
y=np.unique(x)
print(len(y))
```

```python
#task2 find unique tag
Unique_tag = {}

# iterating over the list
for item in mergelist:
   # checking the element in dictionary
   if item in Unique_tag:
      # incrementing the counr
      Unique_tag[item] += 1
   else:
      # initializing the count
      Unique_tag[item] = 1

# printing unique tag with their frequncy
print(Unique_tag)
```

```python
print(len(Unique_tag))
```

```python
#task3 Determine the top-25 Tags appearing frequently
import operator
# Sort Dictionary by value using itemgetter
sorted_tag = dict(sorted(  Unique_tag.items(),
                            key=operator.itemgetter(1),
                            reverse=True))
print('Sorted Dictionary: ')
print(sorted_tag)
```

```python
#Determine the top-25 Tags appearing frequently
from collections import Counter
k = Counter(Unique_tag)
 
# Finding 3 highest values
high = k.most_common(25)
 
 
 
print("top-25 Tags appearing frequently:")
print("Keys: Values")
 
for i in high:
    print(i[0]," :",i[1]," ")
```

```python
#task4-Determine the nature of the distribution of top-500 tag
tag_val=list(sorted_tag.values())
print(tag_val)
```

```python
len(tag_val)
```

```python
distribution=[]
tag_val_r=[]
for i in range(500):
  dis=tag_val[i]-tag_val[i+1]
  distribution.append(dis)
  tag_val_r.append(tag_val[i])
print(len(distribution))
print(distribution)
```

```python
total_tag=list(range(1, 501))
print(len(total_tag))
```

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
x = np.linspace(-2, 2, 10)
y = np.exp(x)
plt.xlabel('500 Tag', fontsize=14)
plt.ylabel('Distribution', fontsize=14)
plt.plot(total_tag, tag_val_r, color='red')
spacing = 0.100
fig.subplots_adjust(bottom=spacing)
plt.show()
```

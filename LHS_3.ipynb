```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#task5 Determine the ratio of duplicate questions asked in each month
dataset_link=pd.read_csv("Arjun_Assignment_data/dataset1/postLinks.csv")
dataset_link
```

```python
#converting into dataframe
data_link=pd.DataFrame(dataset_link)
```

```python
dataset_long=pd.read_csv("Arjun_Assignment_data/dataset1/posts_long.csv")
dataset_long
```

```python
#task1 find no of tag per question
tag_long=dataset_long.values[:,9]
taglong2=list()
tagsl=list()
mergelist=list()
for i in range(len(tag_long)):
  x = tag_long[i].split("<")    
  s = "".join(x)
  x = s.split(">")
  x.pop()
  taglong2.append(x)
  tagsl.append(len(x)) 
  for j in x:
    mergelist.append(j)
```

```python
#task2 find unique tag
Unique_tag = {}

# iterating over the list
for item in mergelist:
   # checking the element in dictionary
   if item in Unique_tag:
      # incrementing the counr
      Unique_tag[item] += 1
   else:
      # initializing the count
      Unique_tag[item] = 1

# printing unique tag with their frequncy
print(Unique_tag)
```

```python
print(len(Unique_tag))
```

```python
#task3 Determine the top-25 Tags appearing frequently
import operator
# Sort Dictionary by value using itemgetter
sorted_tag = dict(sorted(  Unique_tag.items(),
                            key=operator.itemgetter(1),
                            reverse=True))
print('Sorted Dictionary: ')
print(sorted_tag)
```

```python
class my_dictionary(dict):
  
    # __init__ function
    def __init__(self):
        self = dict()
          
    # Function to add key:value
    def add(self, key, value):
        self[key] = value
```

```python
#to find persantage
ans_dict = my_dictionary()
list_dup_val=list(sorted_tagd.values())
list_dup_key=list(sorted_tagd.keys())
list_to_val=list(sorted_tag.values())
list_to_key=list(sorted_tag.keys())
new_val=list()

for i in range(20):
  ans=int((list_dup_val[i]/sorted_tag.get(list_dup_key[i]))*100)
  ans_dict.add(list_dup_key[i],ans)
```

```python
courses = list(ans_dict.keys())
values = list(ans_dict.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='red',
        width = 0.4)
 
plt.xlabel("Tags")
plt.ylabel("Persantage")
plt.show()
  
```
